```python
"""
premise 1...n
conclusion
superordinate if provided
data if provided
group: low/high
property if provided
name if provided
"""

premise_cols = [f'PREMISE CATEGORY {n}' for n in range(1,5)]
useful_cols = premise_cols + ['CONCLUSION CATEGORY', 'SUPERORDINATE CATEGORY', 'PROPERTY', 'DATA', 'GROUP', 'NAME']

col2field = {
    'CONCLUSION CATEGORY': 'conclusion',
    'SUPERORDINATE CATEGORY': 'superordinate',
    'DATA': 'strength',
    'GROUP': 'group',
    'NAME': 'phenomena_name',
    'PROPERTY': 'property'
}
```

```python
def get_induction_data(df):
    # convert df to dict
    relevant_idx = df.columns.intersection(useful_cols)
    df_dict = df[relevant_idx].to_dict('index')
    
    phenomena = []
    
    for idx, mapping in df_dict.items():
        obj = defaultdict(list)
        obj['idx'] = idx+1
        
        for k, v in mapping.items():
            if "PREMISE" in k and v != -99.0:
                obj['premise'].append(v.lower())
            elif v != -99.0:
                if isinstance(v, str):
                    obj[col2field[k]] = v.lower()
                else:
                    obj[col2field[k]] = v
            
        phenomena.append(dict(obj))

    return phenomena
```

```python
get_induction_data(stimuli['Property Type'])
```
